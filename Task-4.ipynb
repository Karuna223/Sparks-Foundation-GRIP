### Decision Tree


 **Explanation:** 
    
    It defined as each node represents a feature/attribute, each branch represents a decision and each leaf represents an outcome(categorical/continues value).
    

**<span style='color:Blue'> Creating the Decision Tree classifier and visualizing it graphically. </span>**

```python
#importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading iris data
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#seperating independent and dependent variables
X = iris.data
y= iris.target
```

```python
Feature_names = iris.feature_names
Target_name = iris.target_names
```

```python
print('Feature names: {}'.format(Feature_names))
print(' ')
print('Target names: {}'.format(Target_name))
```

```python
print('Independent Variables:\n',X)
```

```python
print('Dependent Variable:\n',y)
```

**Now will define the Decision Tree Algorithm**

```python
# Import Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
D_cl = DecisionTreeClassifier()
D_cl.fit(X,y)
```

**visualization through plot_tree**

```python
plt.figure(figsize = (15,10))
plot_tree(D_cl, feature_names = Feature_names, class_names = Target_name, filled = True);
```

<span style='color:Red'> **Understanding the above plot tree**</span>

 **In the first block we have**
    
    petal length<=2.45 
    gini = 0.667
    samples=150
    value=[50,50,50]
    class=setosa
                   

  **How it splits:**

->If tree is less than the petal length score based on result it follows either false or true path

->Gini score is metric that quantifies purity of leaf node
                
        If score is greater than zero that means samples contained within that node belong to 
        different classes(needs further splitting)
                
        If gini score equals to zero that means node is pure/is a leaf node
         
->Samples(150) describe the number of rows in data

->Value([50,50,50]) implies to the target variables setosa,versicolor,virginica

->Class: whichever sample occurs the most within the node will be selected as class value
         

## Splitting The Dataset

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(' ')
print(y_train.shape)
print(y_test.shape)
```

```python
#fit the model in decision classifier
from sklearn.tree import DecisionTreeClassifier
#Make an instance of the Model
clf = DecisionTreeClassifier()
#Train the model on the data
clf = clf.fit(X_train,y_train)
```

```python
#Making predictions
y_pred = clf.predict(X_test)

#Accuracy
from sklearn import metrics
print('Accuracy Score:', metrics.accuracy_score(y_test,y_pred))
```

**The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly**
